# **머신러닝/딥러닝 WorkFlow**

## 1. 문제 정의 (Problem Definition)
- 목표 설정: 무엇을 예측하거나 분류할 것인가? (예: 개/고양이 분류, 집값 예측)
- 유형 결정: 지도 학습(회귀, 분류), 비지도 학습, 강화 학습 중 선택

## 2. 데이터 수집 및 준비 (Data Collection)
- 데이터셋 확보 (CSV, SQL, API, 크롤링 등).
- 판다스(pandas) 등을 사용하여 데이터를 불러오기
- torch는 사용하기 위해서 tensor 사용

## 3. 데이터 전처리 및 탐색 (EDA & Preprocessing)

- EDA(탐색적 데이터 분석): 그래프를 그려 데이터의 분포와 상관관계를 파악
- 결측치 처리: 비어있는 값(NaN)을 채우거나 삭제 (평균값 or 최빈값 등 사용)
- 특성 공학(Feature Engineering): 텍스트를 숫자로 바꾸거나(One-hot encoding), 데이터의 범위를 맞추(Scaling/Normalization).

## 4. 데이터 분할 (Data Splitting)
- 학습용(Train)과 검증/테스트용(Test)으로 데이터를 나누기
- 보통 7:3 또는 8:2 비율 (train_test_split 사용)
- Train은 shuffle=True, Test는 False

## 5. 모델 선택 및 훈련 (Model Selection & Training)
- 문제에 적합한 알고리즘 선택 (Decision Tree, Random Forest, CNN 등).
- model.fit(x_train, y_train) 명령어로 모델을 학습시킵니다.

## 6. 모델 평가 (Evaluation)
- 평가 지표: 정확도(Accuracy), 정밀도(Precision), 재현율(Recall), MSE(회귀의 경우) 등을 확인

## 7. 모델 최적화 및 배포 (Hyperparameter Tuning & Deployment)
- 하이퍼파라미터 튜닝: 학습률이나 나무의 깊이 같은 설정값을 조절해 성능을 극대화합니다.

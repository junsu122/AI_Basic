```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 가상의 데이터 생성 (공부 시간 대비 성적)
X = np.array([1, 2, 3, 4, 5]) # 독립 변수 (공부 시간)
y = np.array([2, 4, 5, 4, 5]) # 종속 변수 (실제 성적)

# 2. 초기화
w = 0.0  # 가중치(Weight, 기울기) 초기값
b = 0.0  # 편향(Bias, 절편) 초기값
learning_rate = 0.01  # 학습률 (보폭)
epochs = 1000         # 반복 횟수

# 3. 경사하강법 실행
n = len(X)

for i in range(epochs):
    # 예측값 계산 (y = wx + b)
    y_pred = w * X + b

    # 손실 함수(MSE) 계산
    loss = np.mean((y_pred - y) ** 2)

    # 미분을 통한 기울기(Gradient) 계산
    dw = (2/n) * np.sum((y_pred - y) * X)
    db = (2/n) * np.sum(y_pred - y)

    # 가중치와 편향 업데이트
    w = w - learning_rate * dw
    b = b - learning_rate * db

    # 100회마다 로그 출력
    if i % 100 == 0:
        print(f"Epoch {i}: Loss {loss:.4f}, w: {w:.4f}, b: {b:.4f}")

# 4. 결과 시각화
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, w * X + b, color='red', label='Regression Line')
plt.title(f"Linear Regression (w={w:.2f}, b={b:.2f})")
plt.legend()
plt.show()
```
